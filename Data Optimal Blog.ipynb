# giving a list of missing values

```python
miss=['na','n/a','NA','-','--']
df=pd.read_csv('D:\Python ML\pandas\data cleaning\property data.csv',na_values=miss)
df
```

```python
print(df['OWN_OCCUPIED'].isnull())
```

# detecting numberss

```python
cnt=0
for row in df["OWN_OCCUPIED"]:
    try:
        int(row)
        df.loc[cnt,'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
df
```

# summarizing missing values

```python
df.isnull().sum()
```

# checking how many NaN values are present in our data

```python
df.isnull().sum().sum()
```

# checking if any undetected values are still remaining

```python
df.isnull().values.any()
```

# filling the median instead of NaN values

```python
df['ST_NUM'].fillna(125,inplace=True)
df
```

# filling median by function

```python
median=df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median,inplace=True)
df
```
